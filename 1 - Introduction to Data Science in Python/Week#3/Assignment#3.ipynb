```python
import numpy as np
import pandas as pd
def answer_one():
    
    energy = pd.read_excel("Energy Indicators.xls", skiprows= 17, skip_footer = 38, parse_cols = "C:F")
    energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    energy['Energy Supply'][energy['Energy Supply'].str.contains("...")==True] = np.nan
    energy['Energy Supply per Capita'][energy['Energy Supply per Capita'].str.contains("...")==True] = np.nan
    
       
    def cleaning_names(index):
        
        index = ''.join([i for i in index if not i.isdigit()])
        index = index.split("(")[0].strip()
        return index
       
    energy['Country'] = energy['Country'].apply(cleaning_names)        
        

    di = {"Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"}
    energy.replace({"Country": di},inplace = True)
    
    energy = energy.set_index("Country")
    print (len(energy.index))
    
    ScimEn = pd.read_excel("scimagojr-3.xlsx")
    ScimEn = ScimEn.set_index("Country")
    print (len(ScimEn.index))
    #del ScimEn.index.name
    
    GPDcolumns = [ "Country Name",'2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    GPD = pd.read_csv("world_bank.csv", skiprows= 4, usecols=GPDcolumns)
    
    di_g = {"Korea, Rep.": "South Korea", 
        "Iran, Islamic Rep.": "Iran",
        "Hong Kong SAR, China": "Hong Kong"}
    GPD.replace({"Country Name": di_g},inplace = True)
    GPD = GPD.set_index("Country Name")
    GPD.index.names = ['Country']
    print (len(GPD.index))
    #del GPD.index.name
    
    result = pd.merge(pd.merge(ScimEn, energy, how='inner', left_index=True, right_index=True), GPD, how='inner', left_index=True, right_index=True)
    result = result.sort_values(["Rank"])
    result_rows = len(result.index)
    print(result_rows)
    result_new = result.head(15)
    del result_new.index.name
    return result_new
#answer_one()
```
